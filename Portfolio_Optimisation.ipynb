### Let's calculate Log Returns instead of Arithmetic Return!

```python
log_return = np.log(stock / stock.shift(1))
```

```python
log_return.head()
```

```python
log_return.hist(bins=100, figsize=(12,8),edgecolor="k")
plt.tight_layout()  #Kernel Density Estimators are better so-to-speak
```

```python
log_return.cov()  #Pairwise co-variance
```

```python
log_return.corr()  #Pairwise co-relation
```

```python
np.random.seed(101)
num_points = 15000
all_weights = np.zeros((num_points,len(stock.columns)))
ret_arr = np.zeros(num_points)
vol_arr = np.zeros(num_points)
sharpe_arr = np.zeros(num_points)

for i in range(num_points):    
       #Create Weights
       weight = np.random.random(4)
       #Normalized Weight
       weight = weight/np.sum(weight)
       all_weights[i][:] = weight  #Numpy BROADCASTING!!  #SAVE IT
       #Expected Return
       ret_arr[i] = np.sum((log_return.mean()*weight)*252)
       #Volatility
       vol_arr[i] = np.sqrt(np.dot(weight.T, np.dot(log_return.cov() * 252, weight)))
       #SR
       sharpe_arr[i] = ret_arr[i]/vol_arr[i]
```

```python
max_pos = sharpe_arr.argmax()
```

```python
max_ret = ret_arr[max_pos]
max_vol = vol_arr[max_pos]
```

```python
# Lets plot out these points
plt.figure(figsize=(12,8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr,cmap='plasma')
plt.colorbar(label="Sharpe Ratio")
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(max_vol, max_ret, color="red", marker="+")
plt.plot()
```

# My own Optimisation Algo

```python
def get_ret_vol_sr(weights):
    weights = np.array(weights)
    ret = np.sum(log_return.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T, np.dot(log_return.cov() * 252, weights)))
    sr = ret/vol
    return np.array([ret,vol,sr])
```

```python
from scipy.optimize import minimize
```

```python
# Now since I wanna maximise sharpe ratio but this minimize function will decrease its value
#so I make it -ve so that when it maximises sharpe ratio it will make its modulus big hence doing my job :D
```

```python
def neg_sharpe(weights):
    return get_ret_vol_sr(weights)[2]*(-1)
```

```python
def check_sum(weights):
    return np.sum(weights) - 1
```

```python
cons = ({'type':'eq','fun': check_sum})
```

```python
bounds = ((0, 1), (0, 1), (0, 1), (0, 1))
```

```python
init_guess = [0.25,0.25,0.25,0.25]
```

```python
#first argument is function u wanna minimize amazing trick here!!
opt_result = minimize(neg_sharpe, init_guess, method="SLSQP", bounds=bounds, constraints=cons)
```

```python
opt_result.x  #these are the weights for MAX SHARPE RATIO <3
```

```python
get_ret_vol_sr(opt_result.x)
```

## Efficient Frontier Plotting

```python
frontier_y = np.linspace(0.1,0.3,100)
```

```python
#Our strategy is that we are gonna optimise the x values i.e. the volatility for all these values
```

```python
#No need to multiply this by -1
def minimize_volatility(weights):
    return get_ret_vol_sr(weights)[1]
```

```python
frontier_volatility = []

for possible_return in frontier_y:
    # function for return
    cons = ({'type':'eq','fun': check_sum},
            {'type':'eq','fun': lambda w: get_ret_vol_sr(w)[0] - possible_return})
    
    result = minimize(minimize_volatility,init_guess,method='SLSQP',bounds=bounds,constraints=cons)
    
    frontier_volatility.append(result['fun'])
```

```python
plt.figure(figsize=(15,10))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap="plasma")
plt.colorbar(label="Sharpe Ratio")
plt.xlabel("Volatility")
plt.ylabel("Returns")
plt.title('MARKOWITZ OPTIMISATION TECHNIQUE')

#Frontier
plt.plot(frontier_volatility, frontier_y, ls="--", lw=3)
plt.show()
```
